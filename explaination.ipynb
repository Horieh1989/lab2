Instead of making a fixed rule (like “always 0 or 1”), we say:

“Look at the new Pokémon. Which known Pokémon is it most similar to?”

That’s what distance measures: similarity.

This way, even if the groups overlap, the algorithm still works by checking neighbors, not a single yes/no cutoff.


Later in machine learning, we often encode labels as numbers:

Pikachu = 1

Pichu = 0

This is just for convenience (computers handle numbers better).
But the decision is still made by checking distances or probabilities, not just the raw 0/1.
